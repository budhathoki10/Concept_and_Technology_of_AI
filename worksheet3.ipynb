3.2 Exploratory Data Analysis Practice Exercise - 1.

```python
#Add a column to the Titanic dataset, fare per year, containing the fare divided by the age of
#the passenger(i.e., Fare/Age).
median_age = dataframe['Age'].median()
dataframe['Age'] = dataframe['Age'].fillna(median_age)
dataframe['fare_per_year']= dataframe['Fare']/dataframe['Age']
print(dataframe.head(6))

#2. Subset rows where fare per year is higher than 5, assigning this to high fare age.
high_fare_age= dataframe[dataframe['fare_per_year']>5]
# print(high_fare_age)

#3. Sort high fare age by descending fare per year, assigning this to high fare age srt.
high_fare_age_srt= high_fare_age.sort_values(by='fare_per_year',ascending=False)
# print(high_fare_age_srt.head(5))

#4. Select only the Name and fare per year columns of high fare age srt and save the result as result.
result = high_fare_age_srt[["Name", "fare_per_year"]]
print(result.head())

#Allison, Master. Hudson Trevor has had the highest fare paid relative to age
```

Which adult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their
class?

```python
#Add a column to the Titanic dataset, fare per year, containing the fare divided by the age ofthe passenger(i.e., Fare/Age).
dataframe['fare_per_class']= dataframe['Fare']/dataframe['Pclass']
# dataframe.head()

# 2. Subset rows where the passenger is male (Sex is ”male”) and an adult (Age is greater than or equalto 18), assigning this to adult males.
adult_males= dataframe[(dataframe["Age"]>=18) & (dataframe["Sex"]=="male")]
# print(adult_males.head())

#3. Sort adult males by descending fare per class, assigning this to adult males srt.
adult_males_srt= adult_males.sort_values(by="fare_per_class",ascending=False)
# print(adult_males_srt.head(6))

#Select only the Name, Age, and fare per class columns of adult males sr and save the results result.
result= adult_males_srt[["Name","Age","fare_per_class"]]
print(result.head(6))
```
